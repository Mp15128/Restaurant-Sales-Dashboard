## 🛠️ Day 4 & 5: Loading & Cleaning
First, let's load our data using Pandas and clean it.

```python
import pandas as pd
df = pd.read_csv("../data/restaurant_sales.csv")
df["Date"] = pd.to_datetime(df["Date"])
df.head()
```

## 📊 Day 6 & 7: Metrics & Visualizations

```python
import matplotlib.pyplot as plt
import seaborn as sns

total_revenue = df["Total"].sum()
print(f"Total Revenue: ${total_revenue:,.2f}")

category_revenue = df.groupby("Category")["Total"].sum()
print(category_revenue)
```

```python
plt.figure(figsize=(8, 5))
category_revenue.plot(kind='bar', color=['#ff6b6b', '#4ecdc4', '#ffe66d'])
plt.title('Revenue by Category')
plt.show()
```
