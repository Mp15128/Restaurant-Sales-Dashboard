## 🧼 Day 5: Data Cleaning & Preprocessing
Before performing metrics calculations, we need to ensure our data has no missing fields, duplicates, or broken formats.

```python
# 1. Check for missing values across all columns
print("Missing Values:\n", df.isnull().sum())
```

```python
# 2. Check for duplicate records
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")

# Drop duplicates if any exist
if duplicate_count > 0:
    df = df.drop_duplicates()
    print("Duplicates dropped successfully.")
```

```python
# 3. Convert 'Date' column from text string to a proper Datetime object
df["Date"] = pd.to_datetime(df["Date"])
print("Date column converted to datetime format successfully!")
print(df["Date"].dtype)
```
