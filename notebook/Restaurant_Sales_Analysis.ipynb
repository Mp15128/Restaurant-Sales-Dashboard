# 🍕 Restaurant Sales Analysis
Welcome to the **Pizza Paradise** analysis notebook. In this project, we will clean, explore, and analyze 1,000 simulated customer orders to extract actionable business insights.

## 🛠️ Day 4: Loading the Dataset
First, let's load our data using Pandas and take a look at its structure.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("../data/restaurant_sales.csv")

# Display the first 5 rows
df.head()
```

```python
# View column data types and structural info
df.info()
```

## 🧼 Day 5: Data Cleaning & Preprocessing
Before performing metrics calculations, we need to ensure our data has no missing fields, duplicates, or broken formats.

```python
# 1. Check for missing values across all columns
print("Missing Values:\n", df.isnull().sum())
```

```python
# 2. Check for duplicate records
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")

# Drop duplicates if any exist
if duplicate_count > 0:
    df = df.drop_duplicates()
    print("Duplicates dropped successfully.")
```

```python
# 3. Convert 'Date' column from text string to a proper Datetime object
df["Date"] = pd.to_datetime(df["Date"])
print("Date column converted to datetime format successfully!")
```

## 📊 Day 6: Business Data Analysis
Now that the data is clean, let's extract key business metrics to see how well our restaurant chain is doing.

```python
# 1. Calculate Total Revenue generated across all stores
total_revenue = df["Total"].sum()
print(f"Total Restaurant Chain Revenue: ${total_revenue:,.2f}")
```

```python
# 2. Calculate the Average Value per customer transaction
avg_order_value = df["Total"].mean()
print(f"Average Transaction Order Value: ${avg_order_value:,.2f}")
```

```python
# 3. Identify the highest selling menu items based on order volume
print("Top Selling Items (by Volume):\n", df["ItemName"].value_counts().head(5))
```

```python
# 4. Group by food category to find out what generates the most revenue
category_revenue = df.groupby("Category")["Total"].sum().sort_values(ascending=False)
print("Revenue Breakdown by Category:\n", category_revenue)
```

```python
# 5. Group by city branches to identify the top performing store location
city_performance = df.groupby("City")["Total"].sum().sort_values(ascending=False)
print("Revenue Breakdown by Branch Location:\n", city_performance)
```

## 📉 Day 7: Data Visualization
Let's turn these tables into visual plots to make the performance trends obvious to anyone looking at our project.

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid")
print("Visualization libraries imported successfully!")
```

```python
# Chart 1: Revenue Breakdown by Category (Bar Chart)
plt.figure(figsize=(8, 5))
category_revenue.plot(kind='bar', color=['#ff6b6b', '#4ecdc4', '#ffe66d'])
plt.title('Total Revenue Breakdown by Menu Category', fontsize=14, fontweight='bold')
plt.ylabel('Revenue ($)')
plt.xlabel('Category')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Chart 2: Top Selling Food Items by Volume (Horizontal Bar Chart)
plt.figure(figsize=(10, 6))
top_items = df["ItemName"].value_counts().head(8)
sns.barplot(x=top_items.values, y=top_items.index, palette="coolwarm")
plt.title('Top 8 Best Selling Items by Volume Order Count', fontsize=14, fontweight='bold')
plt.xlabel('Number of Orders')
plt.ylabel('Menu Item')
plt.tight_layout()
plt.show()
```

```python
# Chart 3: Distribution of Customer Ratings (Histogram)
plt.figure(figsize=(7, 4))
sns.countplot(data=df, x='Rating', palette='viridis')
plt.title('Distribution of Customer Satisfaction Ratings', fontsize=14, fontweight='bold')
plt.xlabel('Star Rating (1-5)')
plt.ylabel('Order Count')
plt.tight_layout()
plt.show()
```
